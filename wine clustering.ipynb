```python
# normalising data using standard scaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dff = scaler.fit_transform(df)
data = pd.DataFrame(dff, columns=df.columns)
data.head()
```

Pca is a dimensionality reducing method which is used to reduce dimensionality of large datasets by transforming large variables into smaller ones without losing any important data.n components is set to 2 to allow visualisation of data in 2D plot.

```python
pca = PCA(n_components=2, n_oversamples=100)
pca_data = pca.fit_transform(data)
# forrming a dataframe from transformed data
data_pca = pd.DataFrame(pca_data, columns=['Pca1', 'Pca2'])
data_pca.head()
```

```python
# visualising transformed data
plt.figure(figsize=(10, 8))
plt.scatter(data_pca.iloc[:, 0], data_pca.iloc[:, 1], s=10)
plt.show()
```

The clusters have not been differentiated due to lack of labels and centroids.

Kmeans clustering is a mehod that aims to partition of n observations in a dataset into a k clusters such that each observation belongs to a specific cluster.

```python
# 
from sklearn.cluster import KMeans
ssd = []
for i in range(2, 10):
    km = KMeans(n_clusters=i)
    Km = km.fit(data_pca)
    labels = km.labels_
    ssd.append(km.inertia_)
```

```python
# plotting an elbow plot which shows the optimal number of clusters to be used in the model.
plt.figure(figsize=(10, 6))
plt.plot(range(2, 10), ssd, 'bx-')
plt.show()
```

from the figure above it shows optimum number of clusters to be at 3.

```python
# using 3 number of clusters in our model
kmeans = KMeans(n_clusters=3)
km = kmeans.fit_predict(data)
labels = kmeans.labels_
# centroid shows the centralpoint of the cluster
centroid = kmeans.cluster_centers_
centroids_df = pca.transform(centroid)


```

```python
# a plot of the clusters
plt.figure(figsize=(10, 6))
plt.scatter(data_pca.iloc[:, 0], data_pca.iloc[:, 1], cmap='rainbow', c=km)
plt.scatter(centroids_df[:, 0], centroids_df[:, 1], marker='x', linewidths=8, color='black')
plt.show()
```

There are three types of clusters that are differentiated by red color,green and purple.
